```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
data = pd.read_csv('savant_data.csv', sep = ',')

data.index = data['game_date']

data.sort_index()
```

```python
plt.figure(figsize = (10, 5))
plot = data['velocity'].sort_index().plot(label = 'Average FB Velocity', color = 'black')
data['velocity'].sort_index().rolling(5).mean().plot(color = 'gold', label = '5-Game Rolling Average')

(data['velocity'].sort_index().rolling(5).mean() - 2 * data['velocity'].sort_index().rolling(5).std()).plot(color = 'grey', linestyle = 'dashed', alpha = 0.5, label = 'Upper and Lower Bounds \n(2 Standard Deviations)')
(data['velocity'].sort_index().rolling(5).mean() + 2 * data['velocity'].sort_index().rolling(5).std()).plot(color = 'grey', linestyle = 'dashed', alpha = 0.5, label = '')

plt.vlines(x = 23.5, ymin = 96.25, ymax = 100.75, color = 'red')
plt.annotate('2025 Season', xy = (24, 100.75), color = 'red')
plt.xticks(rotation = 30)
plt.ylim(96, 101)
plt.xlabel('Game Date')
plt.ylabel('MPH')
plt.title('Paul Skenes Average Fastball Velocity Time Series (2024-2025)')
plt.legend()
plt.show()
```
